# Attract Potential Field

```python
factor = 500/4
xtot = []
ytot = []

xtot1 = []
ytot1 = []

plt.figure(figsize = (8,8))
for file in ['attract1.log','attract2.log','attract3.log','attract4.log']:
    f = open('logs/'+file,"r")
    text = f.read()
    numbers = [float(s) for s in re.findall(r'-?\d+\.?\d*', text)]
    time = numbers[0::14]
    x = numbers[1::14]
    y = numbers[2::14]
    plt.plot(x,y)
    plt.plot(x[0], y[0], 'og')
    plt.plot(x[-1], y[-1], 'ob')
    
    xtot1.append(x)
    ytot1.append(y)
    
    xtot.append((np.array(x)/factor).astype(int))
    ytot.append((np.array(y)/factor).astype(int))
    
for i in range(4):
    xtot[i]=xtot[i][::-1]
    ytot[i]=ytot[i][::-1]
    xtot1[i]=xtot1[i][::-1]
    ytot1[i]=ytot1[i][::-1]
    
plt.show()
```

```python
errors = []
from itertools import product as product

for i in range(15,0,-1):
    dist = 0
    for a in list(product([0,1,2,3],[0,1,2,3])):
        dist += np.abs(xtot1[a[0]][i] - xtot1[a[1]][i]) 
    errors.append(dist)
plt.plot(errors)
```

```python
arrayH = []
for j in range(16):    
    Ni = np.zeros(16)
    for i in range(4):
            Ni[4*ytot[i][j] + xtot[i][j]] += 1 
    pi = process(Ni/(factor**2))
    H = 0
    for p in pi:
        if p != 0:
            H -= p*np.log(p) 
    arrayH.append(H)
plt.plot(arrayH[::-1])
```

# Repulsive potential field

```python
plt.figure(figsize = (8,8))

factor = 500/4
xtot = []
ytot = []

xtot1 = []
ytot1 = []

for file in ['cov1.log','cov2.log','cov3.log','cov4.log']:
    f = open('logs/'+file,"r")
    text = f.read()
    numbers = [float(s) for s in re.findall(r'-?\d+\.?\d*', text)]
    time = numbers[0::14]
    x = numbers[1::14]
    y = numbers[2::14]
    plt.plot(x,y)
    plt.plot(x[0], y[0], 'og')
    plt.plot(x[-1], y[-1], 'ob')
    
    xtot1.append(x)
    ytot1.append(y)
    
    xtot.append((np.array(x)/factor).astype(int))
    ytot.append((np.array(y)/factor).astype(int))

    
plt.show()
```

```python
def process(v):
    norm = np.sum(v)
    if norm == 0: 
         return v
    return v / norm
```

```python
arrayH = []

for i in range(4):
    xtot[i]=xtot[i][::-1]
    ytot[i]=ytot[i][::-1]
    xtot1[i]=xtot1[i][::-1]
    ytot1[i]=ytot1[i][::-1]

for j in range(35):    
    Ni = np.zeros(16)
    for i in range(4):
            Ni[4*ytot[i][j] + xtot[i][j]] += 1 
    pi = process(Ni/(factor**2))
    H = 0
    for p in pi:
        if p != 0:
            H -= p*np.log(p) 
    arrayH.append(H)
plt.plot(arrayH[::-1])
```

```python
errors = []
from itertools import product as product

for i in range(40,5,-1):
    dist = 0
    for a in list(product([0,1,2,3],[0,1,2,3])):
        dist += np.abs(xtot1[a[0]][i] - xtot1[a[1]][i]) 
    errors.append(dist)
plt.plot(errors)
```
